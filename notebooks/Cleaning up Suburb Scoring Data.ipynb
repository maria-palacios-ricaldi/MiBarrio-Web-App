#  keep residential suburbs - drop the rest

```python

# List of suburb names to keep
suburb_names_to_keep = [
    "BANTRY BAY", "CAMPS BAY / BAKOVEN", "CLIFTON", "GREEN POINT", "LLANDUDNO", "MOUILLE POINT",
    "SEA POINT", "BIG BAY", "BLOUBERG", "CENTURY CITY", "MILNERTON", "PARKLANDS", "SUNSET BEACH",
    "TABLE VIEW", "BO-KAAP", "CITY BOWL", "DE WATERKANT", "GARDENS", "OBSERVATORY", "ORANJEZICHT",
    "TAMBOERSKLOOF", "TYGERVALLEY WATERFRONT", "WOODSTOCK", "CLOVELLY", "FISH HOEK", "GLENCAIRN", "KALK BAY",
    "MUIZENBERG", "SIMON'S TOWN", "ST JAMES", "BELLVILLE", "BRACKENFELL", "DURBANVILLE", "EDGEMEAD",
    "GOODWOOD  EXT 1", "GOODWOOD ESTATE", "KRAAIFONTEIN", "PANORAMA", "PAROW", "PLATTEKLOOF", "HOUT BAY", "KOMMETJIE", "NOORDHOEK",
    "SCARBOROUGH", "ATHLONE", "CLAREMONT", "CONSTANTIA", "KENILWORTH", "LAKESIDE", "LANSDOWNE",
    "MOWBRAY", "NEWLANDS", "OTTERY", "PINELANDS", "PLUMSTEAD", "RONDEBOSCH", "ROSEBANK", "TOKAI",
    "WYNBERG", "MELKBOSCH STRAND", "KUILSRIVIER", "KUILS RIVER", "FRESNAYE", "KUILS RIVER"
]

# Convert suburb names in the DataFrame to uppercase
df['Suburb_name'] = df['Suburb_name'].str.upper()

# Filter the DataFrame based on the suburb names
filtered_df = df[df['Suburb_name'].str.contains('|'.join(suburb_names_to_keep), case=False)]

# Initialize a set to store suburb names that have at least one match
matched_suburbs = set()

# Iterate through the suburb names to keep
for subrb in suburb_names_to_keep:
    if any(filtered_df['Suburb_name'].str.contains(suburb, case=False)):
        matched_suburbs.add(suburb)

# Find suburb names that didn't have a match
unmatched_suburbs = set(suburb_names_to_keep) - matched_suburbs
```

```python
# Print the number of rows after filtering
print("Number of rows after filtering:", len(filtered_df))
```

```python
# Save the filtered DataFrame to a CSV file
filtered_df.to_csv('filtered_suburbs_amenity_scores.csv', index=False)
```
